# Data preprocessing 
This notebook contains the code to process the E18 stage scRNA+ATAC mouse embryonic brain dataset. CellRanger output files are available through 10X (https://www.10xgenomics.com/datasets/fresh-embryonic-e-18-mouse-brain-5-k-1-standard-1-0-0)

This dataset was also preprocessed and analyzed in MultiVelo publication (https://www.nature.com/articles/s41587-022-01476-y). See https://github.com/welch-lab/MultiVelo/blob/main/Examples/MultiVelo_Fig2.ipynb for the MultiVelo analysis of this dataset and details of the downloads that are required for both ArchVelo and MultiVelo analysis. Quantified unspliced and spliced counts from Velocyto can be found at the [MultiVelo GitHub page](https://github.com/welch-lab/MultiVelo) and should be downloaded into the ./data/ folder.

Here we re-process this dataset and run scVelo, MultiVelo and ArchVelo on the same set of cells and genes for benchmarking purposes. As a result, we write out adata_rna, adata_atac and adata_atac_raw anndata objects into the processed_data/ folder. adata_rna is used for all analyses; adata_atac is used for MultiVelo analyses; adata_atac_raw is used for ArchVelo analyses.

Please note that MultiVelo and ArchVelo differ significantly in their ATAC modality preprocessing pipeline. The input count matrix for the ArchVelo ATAC analysis (summit_counts_matrix) can be downloaded at https://figshare.com/projects/ArchVelo/229794

Alternatively, the outputs can be downloaded at the same figshare link and this notebook can be omitted.

## Preprocessing the ATAC counts

```python
#read
adata_atac_raw = sc.read_csv('data/outs/our_peaks/summit_counts_matrix.csv').T
#adata_atac_raw = adata_atac_raw[:,adata_atac_raw.var['feature_types'] == "Peaks"]
```

```python
adata_atac_raw.layers['raw_counts'] = adata_atac_raw.X#.todense()
```

```python
min(adata_atac_raw.layers['raw_counts'].astype(bool).sum(0))
```

```python
0.01*adata_atac_raw.shape[0]
```

```python
adata_atac_raw.layers['libnorm'] = (np.array(adata_atac_raw.layers['raw_counts']).T/np.array(adata_atac_raw.layers['raw_counts']).sum(1)).T*np.median(np.array(adata_atac_raw.layers['raw_counts']).sum(1))
```

```python
adata_atac_raw.layers['poisson_corrected'] = np.ceil(adata_atac_raw.layers['raw_counts']/2)
```

```python
adata_atac_raw.shape
```

```python
adata_atac_raw.X = np.array(adata_atac_raw.layers["poisson_corrected"].copy())
sc.experimental.pp.normalize_pearson_residuals(adata_atac_raw, theta=1)
adata_atac_raw.layers["pearson"] = adata_atac_raw.X.copy()
# adata_atac_raw.layers['zscored'] = np.nan_to_num(scipy.stats.zscore(adata_atac_raw.layers['libnorm']))
# adata_atac_raw.layers['zscored_pearson'] = np.nan_to_num(scipy.stats.zscore(adata_atac_raw.layers['pearson']))
```

```python
adata_atac_raw = adata_atac_raw[:, ~np.isnan(adata_atac_raw.layers["pearson"].sum(0))]
```

```python
adata_atac_raw.shape
```

## Finding shared barcodes and features between RNA and ATAC

```python
# #original result
# adata_result = sc.read_h5ad("multivelo_result.h5ad")
# adata_result
```

```python
shared_cells = pd.Index(np.intersect1d(adata_rna.obs_names, adata_atac.obs_names))
shared_genes_atac = pd.Index(np.intersect1d(adata_rna.var_names, adata_atac.var_names))
shared_genes_atac_raw = pd.Index(np.intersect1d(adata_rna.var_names, rel_genes))
print('# cells shared between processed RNA and multivelo filtered ATAC matrix: ',len(shared_cells))
print('# genes shared between RNA var_names and multivelo aggregated ATAC var_names', len(shared_genes_atac))
print('# genes shared between RNA var_names and the genes to which our ATAC peaks map', len(shared_genes_atac_raw))
```

```python
#mapping somewhere
mapped_genes = np.unique(np.unique(peak_annotation_cop['gene'].dropna()))
#present in rna
rna_genes = adata_rna.var_names
#present in adata_atac
aggr_genes = adata_atac.var_names
#present in adata_atac_raw
atac_raw_genes = np.unique(peak_annotation_cop.loc[adata_atac_raw.var_names]['gene'])
```

```python
shared_genes_total = mapped_genes
shared_genes_total = np.intersect1d(shared_genes_total, rna_genes)
shared_genes_total = np.intersect1d(shared_genes_total, aggr_genes)
shared_genes_total = np.intersect1d(shared_genes_total, atac_raw_genes)
```

```python
print('# genes shared between both methods: ', len(shared_genes_total))
```

```python
#leave only corresponding peaks
rel_peaks_total = adata_atac_raw.var_names[peak_annotation_cop.loc[adata_atac_raw.var_names,:]['gene'].isin(shared_genes_total).values]
```

```python
len(rel_peaks_total)
```

### Reload RNA, subset and process

```python
adata_rna = scv.read("velocyto/10X_multiome_mouse_brain.loom", cache=True)
adata_rna.obs_names = [x.split(':')[1][:-1] + '-1' for x in adata_rna.obs_names]
adata_rna.var_names_make_unique()
```

```python
adata_rna.shape
```

```python
adata_rna_atac_total = adata_rna[shared_cells, shared_genes_total]
adata_atac_total = adata_atac[shared_cells, shared_genes_total]
adata_atac_raw_total = adata_atac_raw[shared_cells, rel_peaks_total]
```

```python
adata_rna = adata_rna_atac_total#.copy()
```

```python
adata_atac = adata_atac_total
adata_atac_raw = adata_atac_raw_total
```

```python
adata_rna.shape
```

```python
scv.pp.normalize_per_cell(adata_rna)
scv.pp.log1p(adata_rna)
scv.pp.moments(adata_rna, n_pcs=30, n_neighbors=50)
```

```python
adata_rna.obs['celltype'] = cell_annot.loc[adata_rna.obs_names, 'celltype']
adata_rna.obs['celltype'] = adata_rna.obs['celltype'].astype('category')
```

```python
# Reorder the categories for color consistency with the manuscript.
all_clusters = ['Upper Layer',
                'Deeper Layer',
                'V-SVZ',
                'RG, Astro, OPC',
                'Ependymal cells',
                'IPC',
                'Subplate']
adata_rna.obs['celltype'] = adata_rna.obs['celltype'].cat.reorder_categories(all_clusters)
```

```python
scv.tl.umap(adata_rna)
scv.pl.umap(adata_rna, color='celltype')
```

```python
adata_atac.obsm['X_umap'] = adata_rna.obsm['X_umap']
adata_atac_raw.obsm['X_umap'] = adata_rna.obsm['X_umap']
```

## Smoothing gene aggregated peaks by neighbors for MultiVelo

```python
# Write out filtered cells and prepare to run Seurat WNN --> R script can be found on Github.
adata_rna.obs_names.to_frame().to_csv('filtered_cells.txt', header=False, index=False)
```

```python
# Read in Seurat WNN neighbors.
nn_idx = np.loadtxt("seurat_wnn/nn_idx.txt", delimiter=',')
nn_dist = np.loadtxt("seurat_wnn/nn_dist.txt", delimiter=',')
nn_cells = pd.Index(pd.read_csv("seurat_wnn/nn_cells.txt", header=None)[0])

# Make sure cell names match.
np.all(nn_cells == adata_atac.obs_names)
```

```python
mv.knn_smooth_chrom(adata_atac, nn_idx, nn_dist)
```
